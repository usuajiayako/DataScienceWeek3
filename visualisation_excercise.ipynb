```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# loading the dataset
dataset = pd.read_csv("all_month.csv")

dataset.head()
```

```python
# creating new columns for time and date
dataset["new_time"] = pd.to_datetime(dataset["time"]).dt.time
dataset["new_date"] = pd.to_datetime(dataset["time"]).dt.date
dataset
```

```python
dataset["type"].value_counts()
```

```python
dataset.isnull().sum()
```

```python
# creating place
dataset["new_place"] = dataset["place"].str.split(",").str.get(1)
dataset
```

```python
dataset["month"] = dataset["time"].str.split("-").str.get(1)
dataset
```

```python
dataset["month"].value_counts()
```

```python
# mapping
dataset["month"] = dataset["month"].map({"10": "October", "09": "September"})
dataset
```

```python
dataset["month"].value_counts()
```

```python
# plot a scatter mapbox
fig = px.scatter_mapbox(dataset, "latitude", "longitude", "mag", zoom = 0.5, mapbox_style = "open-street-map", color_continuous_scale = "Reds", range_color = [4.0, 10.0], animation_frame = "month")
fig.show()
```
