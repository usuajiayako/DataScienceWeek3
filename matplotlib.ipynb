```python
# task 1
people = ["Student A", "Student B"]
studentA = np.array([90, 50, 80, 40])
studentB = np.array([75, 45, 60, 95])
x = np.arange(len(studentA))

fig, ax = plt.subplots(1,2)

ax[0].bar(x, studentA, width = 0.3)
ax[1].bar(x, studentB, width = 0.3)
```

```python

fig, ax = plt.subplots()
width = 0.3
s1bars = ax.bar(x - width/2, studentA, width, label = "studentA")
s2bars = ax.bar(x + width/2, studentB, width, label = "studentB")
ax.set_ylim([0,100])
ax.set_xlabel("Excercises")
ax.set_ylabel("Marks %")
ax.set_xticks([0, 1, 2, 3])
ax.set_xticklabels(["ex1", "ex2","ex3","ex4"])
ax.set_title("Marks by excercises")
ax.legend()


for i in range(len(s1bars)):
    s1bars[i].set_linewidth(3.5)
    if studentA[i] < 50:
       s1bars[i].set_edgecolor("red") 

for i in range(len(s2bars)):
    s2bars[i].set_linewidth(3.5)
    if studentB[i] < 50:
       s2bars[i].set_edgecolor("red") 
```

```python
#Shapes
fig, ax = plt.subplots()
ax.set_ylim([0, 20])
ax.set_xlim([-10, 10])
ax.axis("scaled")

start_x = -2.5
start_y = 2
width = 5
height = 6
facecolor = "gray"

# adding gray rectangle
rectangle = plt.Rectangle((start_x, start_y), width, height, fc=facecolor)
ax.add_patch(rectangle)

# adding brown rectangle
rectangle = plt.Rectangle((-0.5, 2), 1, 2, fc="brown")
ax.add_patch(rectangle)

# adding circles
start_x = -1
start_y = 6
radius = 0.5
facecolor = "yellow"
edgecolor = "black"
linewidth = 1

circle = plt.Circle((start_x, start_y), radius, fc=facecolor, ec=edgecolor, lw=linewidth)
ax.add_patch(circle)

circle = plt.Circle((1, 6), radius, fc=facecolor, ec=edgecolor, lw=linewidth)
ax.add_patch(circle)

# adding a triangle
from matplotlib.patches import Polygon
start_xy = np.array([-2.5,8])
end_xy = np.array([2.5, 8])
top_xy = np.array([0, 10])

triangle = Polygon(np.vstack([start_xy, end_xy, top_xy]), color = "red")
ax.add_patch(triangle)

plt.show()
```
